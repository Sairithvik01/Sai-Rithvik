```python
import numpy as np
a=[1,3,4,5]
a1=[3,4,5,6]
b=np.array(a)
b1=np.array(a1)
print("1.lists",a)
print("2.array",b)
print("3.type",type(a))
print("4.type",type(b))
#mulplication
print("5.mulpication",b*2)
#division
print("6.division",b/2)
#poweroff
print("7.poweroff",b**2)
#addition
print("8.addition of two array",b+b1)
#sin,log,exponicial
print("9.sin",np.sin(b))
print("10.log",np.log(b))
print("11.log2",np.log2(b))
print("12.exp",np.exp(b))
```
